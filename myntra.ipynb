```python
from bs4 import BeautifulSoup

html_content = review_page

soup = BeautifulSoup(html_content, 'html.parser')
r = []
c=[]
n=[]
d=[]
for star_wrapper in soup.find_all("div", class_="user-review-reviewTextWrapper"):
    user_comment=c.append(star_wrapper.text)
for star_wrapper in soup.find_all("div", class_="user-review-starWrapper"):
    rating_span = star_wrapper.find("span", class_="user-review-starRating")
    if rating_span:
        r.append(int(rating_span.text.strip()))
for star_wrapper in soup.find_all("div", class_="user-review-left"):
    rating_span = star_wrapper.find_all("span")
    if rating_span:
        n.append(rating_span[0].text.strip())
        d.append(rating_span[1].text.strip())
    

print(r)
print(c)
print(n)
print(d)
```

```python
def lists_to_dict(keys, values1, values2,values3):
    if not (keys and values1 and values2):
        return {}
    
    return {keys[i]: [values1[i], values2[i],values3[i]] for i in range(len(values1))}

dictionary = lists_to_dict(n, r, c,d)
print(dictionary)
```
